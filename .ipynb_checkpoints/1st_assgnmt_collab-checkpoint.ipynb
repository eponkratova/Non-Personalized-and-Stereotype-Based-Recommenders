```python
#Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

#Listing the existing files in the dir
for root, dirs, files in os.walk("."):  
    for filename in files:
        print(filename)
```

```python
movies = pd.read_csv('./files/movie_row.csv')
movies = movies.set_index('Unnamed: 0')
users = pd.read_csv('./files/users_row.csv')
users = users.set_index('Unnamed: 0')
movies.head()
users.head()
```

```python
#Estimating correlation between users. As we have the matrix containing pairwise correlation between all rows.
cor = users.T.corr()
#Visualizing corr matrix
f, ax = plt.subplots(1, figsize=(10, 10))
sns.heatmap(cor, square=True, cmap='RdYlGn', ax=ax)
```

```python
#Checking that the computation is correct as the course organizers offered some tests i.e. to check math, note that the correlation between users 1648 and 5136 is 0.40298, and the correlation between users 918 and 2824 is -0.31706. All correlations should be between -1 and 1, and the diagonal should be all 1's (since they are self-correlations).
cor.loc[[1648],[5136]]
cor.loc[[918],[2824]]
```

```python
#Identifying the top 5 neighbors (the users with the 5 largest, positive correlations) for users 3867 and 89.
user_thirty_cor = cor.loc[[3867]].T.sort_values(3867, ascending=False).drop(3867).head(5); user_thirty_cor
user_eighty_cor = cor.loc[[89]].T.sort_values(89, ascending=False).drop(89).head(5); user_eighty_cor
```

Task: Identify the top 5 neighbors (the users with the 5 largest, positive correlations) for users 3867 and 89. For example, if the target user were #3712, the closest neighbors are 2824 (corr: 0.46291), 3867 (corr: 0.400275), 5062 (corr: 0.247693), 442 (corr: 0.22713), and 3853 (corr: 0.19366). Don't forget to exclude the target user (corr: 1.0) from your possible selections.

Solution: 2492 (0.476683), 3853 (0.464110), 2486 (0.438992), 3712 (0.400275) and 2288 (0.379856) are the top five neighbors for user 3867.
4809 (0.668516), 5136 (0.562449), 860 (0.539066), 5062 (0.525990) and 3525 (0.475495) are the top five neighbors for user 86.

```python
#Filtering to get only our neighborhood. I need to warn that the implementation is rather ugly.
#Getting the analysis done for user 3867.
user_thirty = users.loc[[2492, 3853, 2486, 3712, 2288],:]
#To simplify the next task, I am filling the missing values with 0s.
user_thirty_filled = user_thirty.fillna(0)
calc_thirty = user_thirty_filled.multiply(user_thirty_cor[3867], axis=0).sum()
us1 = user_thirty.loc[2492,:].apply(lambda x: 0 if pd.isnull(x) else 0.476683)
us2 = user_thirty.loc[3853,:].apply(lambda x: 0 if pd.isnull(x) else 0.464110)
us3 = user_thirty.loc[2486,:].apply(lambda x: 0 if pd.isnull(x) else 0.438992)
us4 = user_thirty.loc[3712,:].apply(lambda x: 0 if pd.isnull(x) else 0.400275)
us5 = user_thirty.loc[2288,:].apply(lambda x: 0 if pd.isnull(x) else 0.379856)
us1_df = pd.DataFrame(us1).T
us2_df = pd.DataFrame(us2).T
us3_df = pd.DataFrame(us3).T
us4_df = pd.DataFrame(us4).T
us5_df = pd.DataFrame(us5).T
calc_2_thirty = pd.concat([us1_df, us2_df, us3_df, us4_df, us5_df]).sum()
calc_3 = calc_thirty / calc_2_thirty; calc_3.sort_values(ascending=False).head(3)
```

Task: Compute the predictions for each movie for users 3867 and 89 by taking the correlation-weighted average of the ratings of the top-five neighbors (for each target user) for each movie. Submit 12 values for this assignment as indicated below. You will be submitting the top three movie IDs and the predicted ratings (to three decimal places) for each user. In a real recommender system, we’d be excluding movies the user has already rated, but do not do this here. Indeed, you should look to see what the user’s rating (if any) is for the top-recommended movies.

Solution: Movies 1891 (4.760290), 155 (4.551455) and 122 (4.507637) got the highest prediction.

```python
#Getting the analysis done for user 89.
user_eighty = users.loc[[4809, 5136, 860, 5062, 3525],:]
user_eighty_filled = user_eighty.fillna(0)
calc_eighty = user_eighty_filled.multiply(user_eighty_cor[89], axis=0).sum()
us1 = user_eighty.loc[4809,:].apply(lambda x: 0 if pd.isnull(x) else 0.668516)
us2 = user_eighty.loc[5136,:].apply(lambda x: 0 if pd.isnull(x) else 0.562449)
us3 = user_eighty.loc[860,:].apply(lambda x: 0 if pd.isnull(x) else 0.539066)
us4 = user_eighty.loc[5062,:].apply(lambda x: 0 if pd.isnull(x) else 0.525990)
us5 = user_eighty.loc[3525,:].apply(lambda x: 0 if pd.isnull(x) else 0.475495)
us1_df = pd.DataFrame(us1).T
us2_df = pd.DataFrame(us2).T
us3_df = pd.DataFrame(us3).T
us4_df = pd.DataFrame(us4).T
us5_df = pd.DataFrame(us5).T
calc_2 = pd.concat([us1_df, us2_df, us3_df, us4_df, us5_df]).sum()
calc_3 = calc_eighty / calc_2; calc_3.sort_values(ascending=False).head(3)
```

Solution: Movies 238 (4.894123), 278 (4.882194) and 807 (4.774092) got the highest prediction.

```python
#Getting the normalized prediction for user 3867
user_thrity_target = users.loc[[3867],:].mean(axis=1)
user_thirty_avg = user_thirty.mean(axis=1)
user_thirty_calc = user_thirty.subtract(user_thirty_avg, axis=0).multiply(pd.concat([us1_df, us2_df, us3_df, us4_df, us5_df]), axis=0).sum().divide(calc_2_thirty)
user_thirty_calc.apply(lambda x: x + user_thrity_target).sort_values(3867,ascending=False).head(3)
```

```python
#Getting the normalized prediction for user 89
user_eighty_target = users.loc[[89],:].mean(axis=1)
user_eighty_avg = user_eighty.mean(axis=1)
user_eighty_calc =user_eighty.subtract(user_eighty_avg, axis=0).multiply(pd.concat([us1_df, us2_df, us3_df, us4_df, us5_df]), axis=0).sum().divide(calc_2)
user_eighty_calc.apply(lambda x: x + user_eighty_target).sort_values(89,ascending=False).head(3)
```
