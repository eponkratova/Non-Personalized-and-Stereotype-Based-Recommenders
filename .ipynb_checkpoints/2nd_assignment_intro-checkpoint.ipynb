```python
df = pd.read_csv('./files/Assignment 2.csv')
df.columns
df.head()
```

```python
#Doing some cleansing by removing some extra columns and rows
df_set = df.drop(['Unnamed: 0', 'Unnamed: 11', 'num-attr', 'Unnamed: 13', 'Unnamed: 16', 'Pred1', 'Pred2'], axis=1).drop(df.index[20:]); df_set
```

```python
#Creating the user profile for User 1
user1_profile = df_set.iloc[:,0:10].multiply(df_set["User 1"], axis=0).sum(); user1_profile
user2_profile = df_set.iloc[:,0:10].multiply(df_set["User 2"], axis=0).sum(); user2_profile
#Estimating how likely a user will like the article
pref_user1 = df_set.drop(['User 1', 'User 2'], axis=1) * user1_profile
pref_user1.sum(axis=1).sort_values(ascending=False)
pref_user2 = df_set.drop(['User 1', 'User 2'], axis=1) * user2_profile
pref_user2.sum(axis=1).sort_values(ascending=False)
```

```python
#Noramlizing the values
df_norm = df_set.iloc[:, 0:10].apply(lambda x: x / x.sum(), axis=1).apply(np.sqrt); df_norm
#Creating the user profile for User 1
user1_profile_norm = df_norm.multiply(df_set["User 1"], axis=0).sum(); user1_profile_norm
user2_profile_norm = df_set.iloc[:, 0:10].apply(lambda x: x / x.sum(), axis=1).multiply(df_set["User 2"], axis=0).sum(); user2_profile
#Estimating how likely a user will like the article
pref_user1_norm = df_norm * user1_profile_norm
pref_user1_norm.sum(axis=1).sort_values(ascending=False)
pref_user2_norm =df_norm * user2_profile_norm
pref_user2_norm.sum(axis=1).sort_values(ascending=False)
```

You may have noticed that in our computation an article that had many attributes checked could have more influence on the overall profile than one that had only a few. doc 1 and doc 19 each have five attributes, while doc6, doc7, and doc18 only have 2 attributes each.
We want to explore whether our simple model may be counting these attribute-heavy documents too much. For example, we might conclude that liking doc6 says more about liking baseball (since it is one of only two attributes for the article along with Europe) than liking doc1 says (since doc1 is also about politics, Asia, soccer, and family).
Type in the answers to the following questions as part of the Quiz:

```python
#Calculating the number of times every term occurs in all docs.
inverse_freq = 1 / df_set.iloc[:,0:10].sum(axis=0); inverse_freq
#Estimating how likely a user will like the article
pref_user1_inverse = df_norm * user1_profile_norm * inverse_freq
pref_user1_inverse.sum(axis=1).sort_values(ascending=False)
pref_user2_inverse =df_norm * user2_profile_norm * inverse_freq
pref_user2_inverse.sum(axis=1).sort_values(ascending=False)
```
